```python
# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Create data
x = np.linspace(0, 10, 100)  # Generate 100 points between 0 and 10
y = np.sin(x)  # Compute the sine of each x value

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave', color='b', linestyle='-', linewidth=2)

# Add title and labels
plt.title('Simple Line Plot')
plt.xlabel('x values')
plt.ylabel('sin(x)')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
